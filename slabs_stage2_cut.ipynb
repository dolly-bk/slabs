```python
import numpy as np
from amplify import (
    BinaryPoly,
    BinaryQuadraticModel,
    sum_poly,
    gen_symbols,
    Solver,
    decode_solution,
)
from amplify import Solver
from amplify.client import FixstarsClient
#from amplify.client.ocean import DWaveSamplerClient
from amplify.constraint import less_equal, equal_to, penalty


client = FixstarsClient()
client.token = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"
client.parameters.timeout = 20000  # Timeout is 20 seconds

#client = DWaveSamplerClient()
#client.token = "DEV-8ccde359d4f754f9ba49a624d1b7aa1a0a8df6a0"
#client.solver = "Advantage_system4.1"
#client.num_reads = 100

solver = Solver(client)
i = df.shape[0] #Groups
d = 15 #Days
k = 4 #UNRS
h = 6 #Thickness
x = gen_symbols(BinaryPoly,k,d,h) #Variable x - thickness for UNRS for day
```

```python
#Function for get array of k-values from Excel table
def array_k(ind):
    k_i=[]
    k_i = (df["Диапазон изменения k"][ind])
    if isinstance(k_i, str):
        k_i = (df["Диапазон изменения k"][ind]).split(',')
    else:
        k_i = [df["Диапазон изменения k"][ind]]
    return(k_i)
```

```python
#Function for (2) for m**2
def sum_by_days_for_m_square(group_range, days_range, group_excl):
    name =0 
    index = 0
    for i_ind in range (group_range):
        k_i = array_k(i_ind)
        if i_ind not in group_excl:
            for d_ind in range (days_range[0], days_range[1]):
                for k_ind in k_i:
                    k_ind = int(k_ind)
                    name +=m_new[i_ind][d_ind][k_ind]**2
    return(name)
```

```python
#Function for (2) for m[d]*m[d-1]
def sum_by_days_for_m_add(group_range, days_range, group_excl):
    name =0 
    name_1 = 0
    name_2 = 0
    index1 = 0
    index2 = 0
    for i_ind in range (group_range):
        k_i = array_k(i_ind)
        if i_ind not in group_excl:
            for d_ind in range (days_range[0], days_range[1]):
                name_1 = 0
                name_2 = 0
                for k_ind in k_i:
                    k_ind = int(k_ind)
                    name_1 +=m_new[i_ind][d_ind+1][k_ind]
                    name_2 +=m_new[i_ind][d_ind][k_ind]
                name += (name_1 * name_2)
    return(name)
```

```python
# Create the objective function (2) - part for x 
cost = sum_poly(
    k,
    lambda k_ind: sum_poly(
        d-1,
        lambda d_ind: sum_poly(
            h,
            lambda h_ind: (x[k_ind][d_ind+1][h_ind] - x[k_ind][d_ind][h_ind])**2),
    ),
)
```

```python
#Create array m for m_lidk variables
m = []
m_start = cost.max_index()+1
for i_ind in range (i):
    for d_ind in range(d):
        for k_ind in range(k):
            count = df["Разрядов(L_i)"][i_ind]
            #print(count)
            if count==0:
                count += 1
            m.append(gen_symbols(BinaryPoly, m_start, (count,)))
            m_start += count
```

```python
#Create array m_new for m_idk variables
index = 0
m_new =[]
for i_ind in range(i):
    m_ind = df["Разрядов(L_i)"][i_ind]
    if m_ind ==0:
        m_ind=1
    for d_ind in range(d):
        for k_ind in range(k):
            m_tmp = (sum_poly(m_ind, lambda j: m[index][j]*(2**j)))
            m_new.append(m_tmp)
            index +=1
```
